```python
M1_M2 = pd.merge(M1_Data, M2_Data, on="ID")
M1_M2_M3 = pd.merge(M1_M2, M3_Data, on="ID")
# alternate approach would be to use concat but it creates duplicate columns and extra effort is needed to remove them
#M1_M2_M3 = pd.concat([M1_Data, M2_Data, M3_Data], axis=1)
```

```python
# Insights of people's interest on Elite Party Services
grouped_interests = M1_M2_M3.groupby(M1_M2_M3.Q8)['Q8'].count()
grouped_interests = pd.DataFrame({'Q8': grouped_interests.index, 'counts': grouped_interests.values})
print(type(grouped_interests))
# res_id_to_label = [(-2, 'Very Unlikely'), (-1, 'Unlikely'), (0, 'Undecided'), (1, 'Likely'), (2, 'Very Likely')]
res_id_to_label = {-2:'Very Unlikely', -1:'Unlikely', 0: 'Undecided', 1: 'Likely', 2: 'Very Likely'}
print(grouped_interests)
grouped_interests['likelihood'] = grouped_interests.apply(lambda row: res_id_to_label[row.Q8], axis=1)
print(grouped_interests)
```

```python
# Insights of people's interest on Elite Party Services

#explode = [0, 0, 0.2, 0.2, 0.2] 
fig, ax = plt.subplots(figsize=(12, 7), subplot_kw=dict(aspect="equal"), dpi= 80)
interest_counts = grouped_interests['counts']
likelihoods = grouped_interests['likelihood']
colors_map = {-2: 'red', -1: 'orange', 0: 'yellow', 1: 'blue', 2: 'green'}
colors = grouped_interests['Q8'].apply(lambda val: colors_map[val]).values
plt.pie(interest_counts, autopct='%.0f%%', colors=colors)
# draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
fig = plt.gcf()
plt.title("People Interests")
plt.ylabel("")
plt.xlabel("")
plt.legend(likelihoods, loc='center')
#Hint: Insert code line here to save your DV as an image.
plt.savefig("A3-Likelihood-Python.png")
plt.show()

'''you can see many people are likely to use Elite party services which is good. 
The next majority is people who are undecided with 23%.
We can target people who are undecided by providing extra benefits which require less effort but enough push to make them likely to use.'''
```
